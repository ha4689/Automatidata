```python
# Convert data columns to datetime

df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

**Trip distance**

```python
# Create box plot of trip_distance

plt.figure(figsize=(5,1))
sns.boxplot(x = df['trip_distance'])
plt.title('Trip Distance Boxplot')
plt.show
```

```python
# Create histogram of trip_distance

plt.figure(figsize=(16,5))
ax = sns.histplot(df['trip_distance'],
             bins=range(0,22,1))
ax.set_xticks(range(0,22,1))
plt.title('Trip Distance Histogram')
median = round(df['trip_distance'].median(),1)
mean = round(df['trip_distance'].mean(),1)
plt.axvline(median,color = 'red', linestyle = "--", label = 'median')
plt.axvline(mean,color = 'blue', linestyle = '--',label = 'mean')
plt.legend()
plt.show
```

Examine the unique values in the `passenger_count` column.

```python
df['passenger_count'].value_counts()
```

```python
# Calculate mean tips by passenger_count

mean_tip_by_passenger_count = df.groupby('passenger_count')[['tip_amount']].mean().reset_index()
mean_tip_by_passenger_count
```

```python
# Create bar plot for mean tips by passenger count

plt.figure(figsize = (16,5))
ax = sns.barplot(data = mean_tip_by_passenger_count,
            x = mean_tip_by_passenger_count.index,
            y = 'tip_amount')
ax.axhline(df['tip_amount'].mean(),ls='--',color='red',label='Mean')
ax.axhline(df['tip_amount'].median(),ls='--',color='orange',label='Median')
ax.legend()
plt.title('Mean Tip By Passenger Count Barplot')
plt.show
```

**Create month and day columns**

```python
# Create a month column

df['pickup_month'] = df['tpep_pickup_datetime'].dt.month_name()

# Create a day column

df['pickup_day'] = df['tpep_pickup_datetime'].dt.day_name()

df.head()
```

**Plot total ride count by month**

```python
# Get total number of rides for each month

pickup_monthly = df.groupby('pickup_month')[['VendorID']].count()
pickup_monthly
```

Reorder the results to put the months in calendar order.

```python
# Reorder the monthly ride list so months go in order

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
         'August', 'September', 'October', 'November', 'December']
pickup_monthly = pickup_monthly.reindex(index = month_order).reset_index()
pickup_monthly
```

```python
# Show the index

pickup_monthly.index
```

```python
# Create a bar plot of total rides per month

plt.figure(figsize=(16,5))
sns.barplot(data = pickup_monthly,
            x = 'pickup_month',
           y = 'VendorID')
plt.ylabel('Passenger Per Month')
plt.xlabel('Pickup Month')
plt.title('Passenger Per Pickup Month Barplot')
plt.show
```

**Plot total ride count by day**

Repeat the above process, but now calculate the total rides by day of the week.

```python
# Repeat the above process, this time for rides by day

pickup_day = df.groupby('pickup_day')[["VendorID"]].count()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
pickup_day = pickup_day.reindex(day_order).reset_index()
pickup_day
```

```python
# Create bar plot for ride count by day

plt.figure(figsize = (10,6))
sns.barplot(data = pickup_day,
           x = 'pickup_day',
           y = 'VendorID')
plt.xlabel('Pickup Day')
plt.ylabel('Passenger Per Day')
plt.title('Passenger Per Pickup Day Barplot')
plt.show
```

**Plot total revenue by day of the week**

Repeat the above process, but now calculate the total revenue by day of the week.

```python
# Repeat the process, this time for total revenue by day

revenue_day = df.groupby('pickup_day')[['total_amount']].sum().reindex(day_order).reset_index()
revenue_day
```

```python
# Create bar plot of total revenue by day

plt.figure(figsize = (16,5))
sns.barplot(data = revenue_day,
           x = 'pickup_day',
           y = 'total_amount')
plt.title('Revenue Per Pickup Day Barplot')

plt.show
```

**Plot total revenue by month**

```python
# Repeat the process, this time for total revenue by month

revenue_month = df.groupby('pickup_month')[['total_amount']].sum().reindex(month_order).reset_index()
revenue_month
```

```python
# Create a bar plot of total revenue by month

plt.figure(figsize=(16,5))
sns.barplot(data = revenue_month,
           x = 'pickup_month',
           y = 'total_amount')
plt.title('Revenue Per Pickup Month Barplot')

plt.show
```

**Plot mean trip distance by drop-off location**

```python
# Get number of unique drop-off location IDs

df['DOLocationID'].nunique()
```

```python
# Calculate the mean trip distance for each drop-off location

distance_by_dropoff = df.groupby('DOLocationID')[['trip_distance']].mean()

# Sort the results in descending order by mean trip distance

distance_by_dropoff = distance_by_dropoff.sort_values('trip_distance').reset_index()
distance_by_dropoff
```

```python
# Create a bar plot of mean trip distances by drop-off location in ascending order by distance

plt.figure(figsize = (16,5))
ax = sns.barplot(data = distance_by_dropoff,
          x = distance_by_dropoff.index,
          y = 'trip_distance')          
ax.set_xticks([])
plt.title('Mean Trip Distan By Drop-Off Location')
plt.show
```

**Histogram of rides by drop-off location**

```python
# Check if all drop-off locations are consecutively numbered

df['DOLocationID'].max() - len(set(df['DOLocationID'])) 
```

To eliminate the spaces in the historgram that these missing numbers would create, sort the unique drop-off location values, then convert them to strings. This will make the histplot function display all bars directly next to each other. 

```python
# DOLocationID column is numeric, so sort in ascending order
sorted_dropoffs = df['DOLocationID'].sort_values()

# Convert to string
sorted_dropoffs = sorted_dropoffs.astype('str')

# Plot
plt.figure(figsize=(16,4))

# DOLocationID column is numeric, so sort in ascending order
sorted_dropoffs = df['DOLocationID'].sort_values()

# Convert to string
sorted_dropoffs = sorted_dropoffs.astype('str')

# Plot
sns.histplot(sorted_dropoffs, bins=range(0, df['DOLocationID'].max()+1, 1))
plt.xticks([])
plt.xlabel('Drop-off locations')
plt.title('Histogram of rides by drop-off location', fontsize=16);
sns.histplot(sorted_dropoffs, bins=range(0, df['DOLocationID'].max()+1, 1))
plt.xticks([])
plt.xlabel('Drop-off locations')
plt.title('Histogram of rides by drop-off location', fontsize=16);
```
